```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import pandas as pd

# Carrega os arquivos CSV
df_motor = pd.read_csv("MOTOR.csv")
df_pressao = pd.read_csv("PRESSAO.csv")
df_trifasico = pd.read_csv("TRIFASICO.csv")
df_volume = pd.read_csv("VOLUME.csv")

# Exibe as primeiras linhas de cada
print("✅ MOTOR:")
display(df_motor.head())

print("✅ PRESSÃO:")
display(df_pressao.head())

print("✅ TRIFÁSICO:")
display(df_trifasico.head())

print("✅ VOLUME:")
display(df_volume.head())


```

```python
print(f"Número de leituras (MOTOR): {len(df_motor)}")
print(f"Número de leituras (TRIFÁSICO): {len(df_trifasico)}")
print(f"Número de leituras (PRESSÃO): {len(df_pressao)}")
print(f"Número de leituras (VOLUME): {len(df_volume)}")
# Junta pelos timestamps (inner join = só os que batem exatamente)
df_merged = pd.merge(df_motor, df_trifasico, on='data', how='inner')
df_merged = pd.merge(df_merged, df_pressao, on='data', how='inner')
df_merged = pd.merge(df_merged, df_volume, on='data', how='inner')
# Renomear colunas para evitar conflitos
df_pressao = df_pressao.rename(columns={'sagenerico': 'pressao'})
df_volume = df_volume.rename(columns={'sagenerico': 'volume'})

print(f"Número de leituras (merged): {len(df_merged)}")

# Leituras não sincronizadas (left-only e right-only)
motor_only = df_motor[~df_motor["data"].isin(df_merged["data"])]
trifasico_only = df_trifasico[~df_trifasico["data"].isin(df_merged["data"])]
print(f"Leituras não sincronizadas (MOTOR): {len(motor_only)}")
print(f"Leituras não sincronizadas (TRIFÁSICO): {len(trifasico_only)}")
```

```python
def classificar_estado_vetorizado(df):
    """
    Classifica o estado do motor com base em pdgenerico e correntes a1, a2, a3.

    Args:
        df (pd.DataFrame): DataFrame com colunas 'pdgenerico', 'a1', 'a2', 'a3'

    Returns:
        pd.Series: Série com a classificação do estado do motor
    """
    try:
        # Garantir que as colunas necessárias existem
        required_columns = ['pdgenerico', 'a1', 'a2', 'a3']
        if not all(col in df.columns for col in required_columns):
            missing = [col for col in required_columns if col not in df.columns]
            raise ValueError(f"Colunas faltando no DataFrame: {missing}")

        # Converter pdgenerico para booleano de forma segura
        motor_on = df['pdgenerico'].astype(bool)
        a1, a2, a3 = df['a1'], df['a2'], df['a3']

        # Condições para classificação
        corrente_acima_1 = (a1 > 1) | (a2 > 1) | (a3 > 1)
        corrente_entre_0_e_1 = ((a1 > 0) & (a1 <= 1)) | ((a2 > 0) & (a2 <= 1)) | ((a3 > 0) & (a3 <= 1))
        corrente_zero = (a1 == 0) & (a2 == 0) & (a3 == 0)

        # Inicializar série de estados com dtype object
        estado = pd.Series(np.nan, index=df.index, dtype='object')

        # Atribuir estados
        estado[motor_on & corrente_acima_1] = 'motor_ligado'
        estado[~motor_on & corrente_zero] = 'motor_desligado'
        estado[motor_on & ~corrente_acima_1 & corrente_entre_0_e_1] = 'motor_em_partida'
        estado[(motor_on & corrente_zero) | (~motor_on & corrente_acima_1)] = 'motor_erro'

        # Preencher valores não classificados
        estado = estado.fillna('motor_indefinido')

        return estado

    except Exception as e:
        print(f"Erro na classificação: {str(e)}")
        return None

# Aplicar a classificação
df_merged['estado_motor'] = classificar_estado_vetorizado(df_merged)

# Verificar resultados
if df_merged['estado_motor'].notnull().any():
    # Contagem de cada estado
    print("\nContagem de cada estado do motor:")
    print(df_merged['estado_motor'].value_counts(dropna=False))

    # Filtrar e mostrar apenas os indefinidos
    print("\nDataFrame com registros indefinidos:")
    df_indefinidos = df_merged[df_merged['estado_motor'] == 'motor_indefinido']
    print(df_indefinidos[['pdgenerico', 'a1', 'a2', 'a3', 'estado_motor']])

    # Informações adicionais sobre indefinidos
    print(f"\nTotal de registros indefinidos: {len(df_indefinidos)}")
    print(f"Percentual de indefinidos: {(len(df_indefinidos)/len(df_merged)*100):.2f}%")
else:
    print("Nenhum estado foi classificado. Verifique o DataFrame de entrada.")



```
